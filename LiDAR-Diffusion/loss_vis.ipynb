```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取csv文件
df = pd.read_csv('metrics_1.csv')

for column in df.columns[:4]:
    plt.plot(df.index[1:101], df[column].iloc[1:101], label=column)

# 添加图例
plt.legend()

# 设置标题和标签
plt.title('Loss Line Plot')
plt.xlabel('kSteps')
plt.ylabel('Losss')

# 显示图形
plt.show()
```
