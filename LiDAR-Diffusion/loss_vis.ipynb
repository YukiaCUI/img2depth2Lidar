```python
import pandas as pd
import matplotlib.pyplot as plt

# csv
df = pd.read_csv('metrics_1.csv')

# 
plt.figure(figsize=(10, 6))
for column in df.columns[:4]:
    # 
    data = df[[column]].dropna()
    plt.plot(data.index[1:101], data[column].iloc[1:101], label=column)

# 
plt.legend()
plt.title('Loss Line Plot')
plt.xlabel('kSteps')
plt.ylabel('Loss')

# 
plt.show()

# 1112
plt.figure(figsize=(10, 6))
for column in df.columns[10:12]:
    # 
    data = df[[column]].dropna()
    plt.plot(data.index[1:101], data[column].iloc[1:101], label=column)

# 
plt.legend()
plt.title('Loss Line Plot (Columns 11 and 12)')
plt.xlabel('kSteps')
plt.ylabel('Loss')

# 
plt.show()
```
