```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取csv文件
df = pd.read_csv('metrics_1.csv')

# 绘制前四列数据的折线图
plt.figure(figsize=(10, 6))
for column in df.columns[:4]:
    # 跳过空值的行
    data = df[[column]].dropna()
    plt.plot(data.index[1:101], data[column].iloc[1:101], label=column)

# 添加图例、标题和标签
plt.legend()
plt.title('Loss Line Plot')
plt.xlabel('kSteps')
plt.ylabel('Loss')

# 显示第一个图形
plt.show()

# 绘制第11和第12列数据的折线图
plt.figure(figsize=(10, 6))
for column in df.columns[10:12]:
    # 跳过空值的行
    data = df[[column]].dropna()
    plt.plot(data.index[1:101], data[column].iloc[1:101], label=column)

# 添加图例、标题和标签
plt.legend()
plt.title('Loss Line Plot (Columns 11 and 12)')
plt.xlabel('kSteps')
plt.ylabel('Loss')

# 显示第二个图形
plt.show()
```
